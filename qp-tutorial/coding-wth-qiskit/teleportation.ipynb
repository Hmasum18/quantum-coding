```python
from qiskit import *
from qiskit.visualization import *
```

```python
# make a quantum circuit with 3 qubit and 3 classical register to measure them
circuit = QuantumCircuit(3, 3)
%matplotlib inline

circuit.draw(output="mpl")
```

```python
# apply an XNOT gate to qubit 0
circuit.x(0)
circuit.barrier()
circuit.draw(output='mpl')
```

```python
# apply an H-gate on qubit-1
circuit.h(1)

# apply an CNOT gate to qubit 1 and 2 to create entanglement
circuit.cx(1,2)

# visualize the circuit
circuit.draw(output="mpl")
```

```python
# apply an CNOT gate to qubit 0 and 1
circuit.cx(0,1)

# apply and H-gate to qubit 0
circuit.h(0)

circuit.draw(output="mpl")
```

```python
circuit.barrier()
# measure qubit 0 and 1 in classicla register 0 and 1
circuit.measure([0,1] , [0,1])

circuit.draw(output='mpl')
```

```python
circuit.barrier()

circuit.cx(1, 2)
circuit.cz(0, 2)

circuit.measure(2,2)

circuit.draw(output="mpl")
```

```python
simulator = Aer.get_backend('qasm_simulator')
result = execute(circuit, backend=simulator, shots=1024).result()
counts = result.get_counts()
plot_histogram(counts)
```
